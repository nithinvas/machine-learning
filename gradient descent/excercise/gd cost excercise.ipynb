```python
df =pd.read_csv('C:\\Workspace\\machine learning\\gradient descent\\excercise\\marks.csv')
df
```

```python
x_train=df['math'].to_numpy()
y_train=df['cs'].to_numpy()
print(x_train)
print(y_train)
```

```python
def grad_descent(x,y):
    mcur=0
    bcur=0
    iterations=1000
    learning_rate=0.002
    n=len(x)
    for i in range(iterations):
        y_predicted=mcur*x+bcur
        cost=(1/n)*sum((y-y_predicted)*(y-y_predicted))
        md=-(2/n)*sum(x*(y-y_predicted))
        bd=-(2/n)*sum(y-y_predicted)
        mcur=mcur-learning_rate*md
        bcur=bcur-learning_rate*bd
        print("m {}, b {}, iteration {}, cost {}".format(mcur,bcur,i,cost))
        
grad_descent(x_train,y_train)
    
```
